```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Define recession periods
recession_years = [1990, 1991, 2008, 2009]

# Create a new column to indicate recession/non-recession
df['Recession'] = df['Year'].apply(lambda year: 'Recession' if year in recession_years else 'Non-Recession')

# Filter data for recession periods
recession_data = df[df['Recession'] == 'Recession']

# Map vehicle types to price categories
price_mapping = {
    'Supperminicar': 1,  # Low
    'Smallfamiliycar': 2,  # Medium
    'Mediumfamilycar': 3   # High
}
recession_data['Price_Category'] = recession_data['Vehicle_Type'].map(price_mapping)

# Calculate average price category and sales for each year
yearly_data = recession_data.groupby('Year').agg(
    Average_Price_Category=('Price_Category', 'mean'),
    Average_Sales_Volume=('Automobile_Sales', 'mean')
)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(yearly_data['Average_Price_Category'], yearly_data['Average_Sales_Volume'], alpha=0.7)

# Customize the chart
plt.title('Correlation between Average Vehicle Price and Sales Volume during Recessions')
plt.xlabel('Average Vehicle Price Category (1=Low, 2=Medium, 3=High)')
plt.ylabel('Average Sales Volume')
plt.grid(True)

plt.show()
```

```python
import folium
import pandas as pd

# Assuming 'df' is your DataFrame

# Define recession periods
recession_years = [1990, 1991, 2008, 2009]

# Create a new column to indicate recession/non-recession
df['Recession'] = df['Year'].apply(lambda year: 'Recession' if year in recession_years else 'Non-Recession')

# Filter data for recession periods
recession_data = df[df['Recession'] == 'Recession']

# **Important**: You'll need columns in your DataFrame that represent:
#   - 'Region' or 'Office' (for grouping)
#   - Latitude and Longitude of those regions/offices

# If 'Region' column is not present, create a dummy 'Region' column for demonstration:
# (Replace with your actual region data if available)
recession_data['Region'] = np.random.choice(['North', 'South', 'East', 'West'], size=len(recession_data))

# Example: Assuming your DataFrame now has 'Region', 'Latitude', 'Longitude' columns
# Group by region and sum sales during recession
sales_by_region = recession_data.groupby('Region')['Automobile_Sales'].sum().reset_index()

# Merge sales data with region coordinates
# (assuming you have a separate DataFrame 'region_coords' with Region, Latitude, Longitude)
region_coords = pd.DataFrame({  # Example coordinates (replace with your data)
    'Region': ['North', 'South', 'East', 'West'],
    'Latitude': [40.7128, 25.7617, 34.0522, 37.7749],
    'Longitude': [-74.0060, -80.1918, -118.2437, -122.4194]
})
sales_with_coords = pd.merge(sales_by_region, region_coords, on='Region')

# Create the map
m = folium.Map(location=[37.7749, -122.4194], zoom_start=5)  # Center on average US location

# Add markers for each region with sales as popup
for index, row in sales_with_coords.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Region: {row['Region']}<br>Sales: {row['Automobile_Sales']}",
        icon=folium.Icon(color="red", icon="info-sign")
    ).add_to(m)

# Display the map
display(m)
```
