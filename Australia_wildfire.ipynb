```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the wildfire data

# Group data by year and calculate the average estimated fire area
average_fire_area_by_year = df.groupby('Year')['Estimated_fire_area'].mean()

# Create the plot
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
average_fire_area_by_year.plot(kind='line', marker='o')
# kind='line' creates a line plot, marker='o' adds circular markers to data points

plt.xlabel('Year')
plt.ylabel('Average Estimated Fire Area (km²)')
plt.title('Change in Average Estimated Fire Area Over Time')
plt.grid(True)  # Add a grid for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with wildfire data and 'Year' and 'Month' columns

# Filter data for the years 2010 to 2013
filtered_df = df[(df['Year'] >= 2010) & (df['Year'] <= 2013)]

# Group data by year and month, and calculate the average estimated fire area
average_fire_area_by_year_month = filtered_df.groupby(['Year', 'Month'])['Estimated_fire_area'].mean().reset_index()

# Create the plot
plt.figure(figsize=(14, 7))
for year in range(2010, 2014):
    year_data = average_fire_area_by_year_month[average_fire_area_by_year_month['Year'] == year]
    plt.plot(year_data['Month'], year_data['Estimated_fire_area'], label=year, marker='o')

plt.xlabel('Month')
plt.ylabel('Average Estimated Fire Area (km²)')
plt.title('Average Estimated Fire Area by Month (2010-2013)')
plt.xticks(range(1, 13))  # Set x-axis ticks to represent months
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Creating a bar plot using seaborn to visualize the distribution of mean estimated fire brightness across regions
plt.figure(figsize=(10, 6))
# Using seaborn's barplot function to create the plot
sns.barplot(data=df, x='Region', y='Mean_estimated_fire_brightness')
plt.xlabel('Region')
plt.ylabel('Mean Estimated Fire Brightness (Kelvin)')
plt.title('Distribution of Mean Estimated Fire Brightness across Regions')
plt.show()
```

```python
# Creating a pie chart to visualize the portion of count of pixels for presumed vegetation fires across regions
plt.figure(figsize=(10, 6))
# Grouping the data by region and summing the counts
region_counts = df.groupby('Region')['Count'].sum()
# Creating the pie chart using plt.pie function
# Labels are set to the region names, and autopct is used to display percentage
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%')
plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region')
plt.axis('equal')
plt.show()
```

```python

# Creating a pie chart to visualize the portion of count of pixels for presumed vegetation fires across regions
plt.figure(figsize=(10, 6))

# Grouping the data by region and summing the counts
region_counts = df.groupby('Region')['Count'].sum()

# Creating the pie chart without autopct
plt.pie(region_counts, labels=region_counts.index)  # Remove autopct

# Set title
plt.title('Percentage of Pixels for Presumed Vegetation Fires by Region')

# Add legend with calculated percentages
plt.legend([(i, round(k / region_counts.sum() * 100, 2))
            for i, k in zip(region_counts.index, region_counts)],
           loc="best")  # Adjust loc for optimal placement

plt.axis('equal')
plt.show()
```

```python
@app.callback([Output(component_id='plot1', component_property='children'),
               Output(component_id='plot2', component_property='children')],
              [Input(component_id='region', component_property='value'),
               Input(component_id='year', component_property='value')])
def update_graph(region_chosen, year_chosen):
    # Define the function to update the graphs based on region and year input

    # Filter the dataframe
    filtered_df = df[(df['Region'] == region_chosen) & (df['Year'] == year_chosen)]

    # Create the first plot (e.g., a bar chart)
    fig1 = px.bar(filtered_df, x='Month', y='Estimated_fire_area', title=f'Estimated Fire Area in {region_chosen} for {year_chosen}')

    # Create the second plot (e.g., a scatter plot)
    fig2 = px.scatter(filtered_df, x='Mean_estimated_fire_brightness', y='Mean_estimated_fire_radiative_power',
                      title=f'Brightness vs. Radiative Power in {region_chosen} for {year_chosen}')

    return [dcc.Graph(figure=fig1), dcc.Graph(figure=fig2)]  # Return the graph components
```

```python
from jupyter_dash import JupyterDash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Sample Data
df = px.data.gapminder()

# Initialize Dash App
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("Interactive Dashboard in Google Colab"),
    dcc.Dropdown(
        id='continent-dropdown',
        options=[{'label': i, 'value': i} for i in df['continent'].unique()],
        value='Asia'
    ),
    dcc.Graph(id='line-chart')
])

# Callback for Interactivity
@app.callback(
    dash.Output('line-chart', 'figure'),
    [dash.Input('continent-dropdown', 'value')]
)
def update_graph(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    fig = px.line(filtered_df, x="year", y="gdpPercap", color="country")
    return fig
```
