```python
import requests
import pandas as pd
from pandas import json_normalize

# API Endpoints
urls = {
    "launches": "https://api.spacexdata.com/v4/launches/past",
    "rockets": "https://api.spacexdata.com/v4/rockets",
    "launchpads": "https://api.spacexdata.com/v4/launchpads",
    "payloads": "https://api.spacexdata.com/v4/payloads",
    "cores": "https://api.spacexdata.com/v4/cores"
}

# Fetch data from all endpoints
dfs = {}  # Dictionary to store DataFrames
for key, url in urls.items():
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        dfs[key] = json_normalize(data)
        print(f"{key} data successfully fetched and converted to DataFrame.")
    else:
        print(f"Error fetching {key} data.")

# Display first few rows of each dataset
for key, df in dfs.items():
    print(f"\nFirst rows of {key} dataset:")
    display(df.head())
```

```python
df_launches = dfs["launches"]
df_rockets = dfs["rockets"]
df_launchpads = dfs["launchpads"]
df_payloads = dfs["payloads"]
df_cores = dfs["cores"]

# Select useful columns from rockets
df_rockets = df_rockets[['id', 'name', 'stages', 'cost_per_launch', 'success_rate_pct']]
df_rockets.rename(columns={'id': 'rocket_id', 'name': 'rocket_name'}, inplace=True)

# Select useful columns from launchpads
df_launchpads = df_launchpads[['id', 'name', 'locality', 'region']]
df_launchpads.rename(columns={'id': 'launchpad_id', 'name': 'launchpad_name'}, inplace=True)
```

```python
# Merge launches with rockets to get rocket details
df_launches = df_launches.merge(df_rockets, left_on='rocket', right_on='rocket_id', how='left')

# Merge launches with launchpads to get location details
df_launches = df_launches.merge(df_launchpads, left_on='launchpad', right_on='launchpad_id', how='left')

# Drop unnecessary ID columns
df_launches.drop(columns=['rocket', 'launchpad', 'rocket_id', 'launchpad_id'], inplace=True)

# Display the final cleaned dataset
df_launches.head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
df = pd.read_csv("dataset_part_2.csv")  # Replace with actual dataset

# Step 1: Inspect Data
print(df.info())
print(df.head())

# Step 2: Handle Missing Values
df.fillna({"LandingPad": "Unknown"}, inplace=True)

# Step 3: Success Rate per Launch Site
success_rate = df.groupby("LaunchSite")["Class"].mean() * 100
print("\nSuccess Rate by Launch Site:")
print(success_rate)

# Step 4: Plot Payload Mass vs. Landing Success
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Class", y="PayloadMass")
plt.xlabel("Landing Success (0 = Fail, 1 = Success)")
plt.ylabel("Payload Mass (kg)")
plt.title("Payload Mass vs. Landing Success")
plt.show()

# Step 5: One-Hot Encoding for Categorical Features
categorical_features = ["LaunchSite", "Orbit", "BoosterVersion"]
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[categorical_features]).toarray()

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out())

# Step 6: Merge Encoded Features
df.drop(columns=categorical_features, inplace=True)  # Drop original categorical columns
df = pd.concat([df, encoded_df], axis=1)

# Step 7: Display Processed Data
print(df.head())

# Save processed dataset for ML
df.to_csv("falcon9_cleaned.csv", index=False)
```

```python
import requests
import pandas as pd
import io
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC

# Download dataset part 2
URL1 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
response = requests.get(URL1)

if response.status_code == 200:
    data_part_2 = pd.read_csv(io.BytesIO(response.content))
else:
    print(f"Failed to download data. Status code: {response.status_code}")

# Download dataset part 3
URL2 = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_3.csv'
response = requests.get(URL2)

if response.status_code == 200:
    data_part_3 = pd.read_csv(io.BytesIO(response.content))
else:
    print(f"Failed to download data. Status code: {response.status_code}")

# Combine the data if needed (you can adjust this step depending on how you want to merge them)
data = pd.concat([data_part_2, data_part_3], axis=1)

# Check the first few rows of the combined data
print(data.head())

# Assume 'Class' is the column we are predicting.
# Prepare features (X) and target (y)
X = data.drop('Class', axis=1)  # Replace 'target' with 'Class'
y = data['Class']  # Replace 'target' with 'Class'

# Convert the 'Date' column to datetime objects if it's not already
X['Date'] = pd.to_datetime(X['Date'])

# Extract year from the 'Date' column and create a new feature 'Year'
X['Year'] = X['Date'].dt.year

# Drop the original 'Date' column
X = X.drop('Date', axis=1)

# One-hot encode categorical features (if any)
categorical_features = X.select_dtypes(include=['object']).columns
if len(categorical_features) > 0:
    encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for compatibility with StandardScaler
    encoded_features = encoder.fit_transform(X[categorical_features])
    encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))
    X = X.drop(categorical_features, axis=1)
    X = pd.concat([X, encoded_df], axis=1)

# Step 1: Standardize the Data
transform = StandardScaler()
X_scaled = transform.fit_transform(X)

# Step 2: Split the Data into Training and Testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 3: Split Training Data into Training and Validation Sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Step 4: Hyperparameter Tuning using GridSearchCV
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}  # Example grid for SVM
grid_search = GridSearchCV(SVC(), param_grid, cv=5)  # 5-fold cross-validation
grid_search.fit(X_train, y_train)

# Best Parameters and Best Score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

# Evaluate the model on the test set
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test score:", test_score)
```
